```python
#################################################################################
# Генетические алгоритмы. Почитать про применение Г.А. в решении                #
# уравнений/диф.ур. Выбрать простой класс уравнений и для него воспроизвести    #
# результат. Т.е. построить решатель с помощью Г.А. Придумать метрику и         #
# проверить качество решателя                                                   #
#################################################################################

import numpy as np
import matplotlib.pyplot as plt
from geneticalgorithm import geneticalgorithm as ga

# Правая часть ОДУ
def F1(t, x):
    return x / t + t;

# Отклонение в одной точке дробления
def dotError(F, T, X, i):
    der = (X[i + 1] - X[i]) / (T[i + 1] - T[i])
    return (der - F(T[i], X[i])) ** 2

# Отклонение на всем дроблении
def Error(F, T, X):
    error = 0
    for i in range(T.size - 1):
        error += dotError(F, T, X, i)
    return error

# Дробление промежутка
T = np.arange(0.1, 1.1, 0.1)
X_etalon = np.arange(0.1, 1.1, 0.1)

# Функция приспособленности особей
fitness = lambda X : Error(F1, T, X)

for i in range(T.size):
    X_etalon[i] = (T[i] + 0.9) * T[i]

varbound=np.array([[0, 2]] * 10)
algorithm_param = {'max_num_iteration': 500,
                   'population_size':200,
                   'mutation_probability':0.04,
                   'elit_ratio': 0.01,
                   'crossover_probability': 0.5,
                   'parents_portion': 0.25,
                   'crossover_type':'uniform',
                   'max_iteration_without_improv':None}

model=ga(function=fitness, 
         dimension=10, 
         variable_type='real', 
         variable_boundaries=varbound,
         algorithm_parameters=algorithm_param)

model.run()

plt.rcParams['figure.figsize'] = [8, 8]
plt.title('Genetic algorithm')
plt.axis([0, 1.1, 0, 2.1])
plt.plot(T, X_etalon, 'blue')
plt.plot(T, model.output_dict['variable'], 'red')
plt.legend(['Original', 'Approximation'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from geneticalgorithm import geneticalgorithm as ga

# Правая часть ОДУ
def F2(t, x):
    return math.cos(t);

# Отклонение в одной точке дробления
def dotDifError(F, T, X, i):
    der = (X[i + 1] - X[i]) / (T[i + 1] - T[i])
    return (der - F(T[i], X[i])) ** 2

def dotDerError(T, X, i):
    der1 = (X[i + 1] - X[i]) / (T[i + 1] - T[i])
    der2 = (X[i] - X[i - 1]) / (T[i] - T[i - 1])
    return (der1 - der2) ** 2

# Отклонение на всем дроблении
def Error(F, T, X):
    error = 0
    for i in range(T.size - 1):
        error += dotDifError(F, T, X, i)
    error = math.sqrt(error)
    for i in range(T.size - 2):
        error += dotDerError(T, X, i + 1)
    return error

# Дробление промежутка
T = np.arange(0, 3.24, 0.1)
X_etalon = np.arange(0, 3.24, 0.1)

# Функция приспособленности особей
fitness = lambda X : Error(F2, T, X)

for i in range(T.size):
    X_etalon[i] = math.sin(T[i])

varbound=np.array([[-1.1, 1.1]] * 33)
algorithm_param = {'max_num_iteration': 1000,
                   'population_size':150,
                   'mutation_probability':0.05,
                   'elit_ratio': 0.01,
                   'crossover_probability': 0.5,
                   'parents_portion': 0.25,
                   'crossover_type':'uniform',
                   'max_iteration_without_improv':None}

model=ga(function=fitness, 
         dimension=33, 
         variable_type='real', 
         variable_boundaries=varbound,
         algorithm_parameters=algorithm_param)

model.run()

plt.rcParams['figure.figsize'] = [8, 8]
plt.title('Genetic algorithm')
plt.axis([0, 6.5, -1.1, 1.1])
plt.plot(T, X_etalon, 'blue')
plt.plot(T, model.output_dict['variable'], 'red')
plt.legend(['Original', 'Approximation'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from geneticalgorithm import geneticalgorithm as ga

# Правая часть ОДУ #1
def F1(t, x):
    return math.cos(t);

# Правая часть ОДУ #2
def F2(t, x):
    return x ** 2 + 1;

# Правая часть ОДУ #3
def F3(t, x):
    return x / t + t;

# Расчитывает значение полинома в точке
def evOrigin(polinom, t):
    value = 0
    for i in range(polinom.size):
        value += (t ** i) * polinom[i]
    return value

# Расчитывает значение производной полинома в точке
def evDerivative(polinom, t):
    value = 0
    for i in range(polinom.size - 1):
        value += (i + 1) * (t ** i) * polinom[i + 1]
    return value

# Функция ошибки
def error(F, T, t0, x0, polinom):
    error = 0
    for i in range(T.size):
        error += (evDerivative(polinom, T[i]) - F(T[i], evOrigin(polinom, T[i]))) ** 2
    error += (evOrigin(polinom, t0) - x0) ** 2
    return error

# Полезные константы
POLINOM_DEG = 7
N_GENERATION = 1000

# Дробление промежутка
T = np.arange(-1.5, 1.6, 0.1)
polinom = np.zeros(POLINOM_DEG)
etalonX = np.zeros(T.size)
approxX = np.zeros(T.size)

# Функция приспособленности особей
fitness = lambda P : error(F2, T, 0, 0, P)

# Инициализируем параметры генетического алгоритма
varbound=np.array([[-1, 1]] * POLINOM_DEG)
algorithm_param = {'max_num_iteration': N_GENERATION,
                   'population_size':100,
                   'mutation_probability':0.07,
                   'elit_ratio': 0.01,
                   'crossover_probability': 0.5,
                   'parents_portion': 0.2,
                   'crossover_type':'uniform',
                   'max_iteration_without_improv':None}
model=ga(function=fitness, 
         dimension=POLINOM_DEG, 
         variable_type='real', 
         variable_boundaries=varbound,
         algorithm_parameters=algorithm_param)

# Запускаем генетический алгоритм
model.run()

# Находим значения в точках
for i in range(T.size):
    approxX[i] = evOrigin(model.output_dict['variable'], T[i])
for i in range(T.size):
    etalonX[i] = math.tan(T[i])

# Рисуем график
fig, ax = plt.subplots()
fig.set_figwidth(8)
fig.set_figheight(8)
ax.set_title('Genetic Algorithm Approximation')
ax.axis([-3, 3, -5, 5])
ax.plot(T, etalonX, 'blue')
ax.plot(T, approxX, 'red')
ax.legend(['Original', 'Approximation'])
ax.grid(linestyle = '--')
plt.show()
```
