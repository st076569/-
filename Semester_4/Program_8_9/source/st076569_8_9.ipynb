```python
#################################################################################
# Алгоритм муравьиной колонии. Поэкспериментировать с задачей коммивояжера.     #
# Т.е. решить задачу коммивояжера с разными наборами  настраиваемых параметров. #
# Провести анализ результата.                                                   #
#################################################################################
# Метод имитации отжига. Поэкспериментировать с задачей коммивояжера. Т.е.      #
# решить задачу коммивояжера с разными наборами настраиваемых параметров.       #
# Провести анализ результата. Сравнить результаты с 8-ой задачей.               #
# Проанализировать глобальную оптимальность решений.                            #
#################################################################################

# Импортируем нужные библиотеки
import numpy as np
import matplotlib.pyplot as plt
from satsp import solver
from neataco import AntColonyOptimization

# Муравьиный алгоритм
def antColony(x, y, ant_count=500, alpha=1, beta=2, evaporation=0.2):
    # Подготовка данных для дальнейшей оптимизации
    num = np.array([i for i in range(x.size)])
    num = num[:, np.newaxis]
    x1 = x[:, np.newaxis]
    y1 = y[:, np.newaxis]
    items = np.hstack([num, x1, y1])
    
    # Подготовка муравьиного алгоритма
    distance = lambda a, b: np.sqrt((a[1] - b[1]) ** 2 + (a[2] - b[2]) ** 2)
    aco = AntColonyOptimization(items, distance, ant_count, 
                                alpha, beta, evaporation)
    optimal_path = aco.run()
    new_x = np.zeros(x.size)
    new_y = np.zeros(y.size)
    
    # Перестановка координат в соответствии с оптимальным обходом городов
    for i in range(x.size):
        new_x[i] = x[int(optimal_path[i][0])]
        new_y[i] = y[int(optimal_path[i][0])]
    
    # Возвращаем обновленную последовательность координат
    return new_x, new_y

# Метод имитации отжига
def simulatedAnnealing(x, y, t_beg=100, t_end=1, cooling=0.99):
    # Подготовка данных для дальнейшей оптимизации
    num = np.array([i for i in range(x.size)])
    num = num[:, np.newaxis]
    x1 = x[:, np.newaxis]
    y1 = y[:, np.newaxis]
    items = np.hstack([num, x1, y1])
    
    # Подготовка метода имитации отжига
    solver.Solve(items, start_temp=t_beg, stop_temp=t_end, 
                 alpha=cooling, screen_output=False)
    optimal_path = solver.GetBestTour()
    new_x = np.zeros(x.size)
    new_y = np.zeros(y.size)
    
    # Перестановка координат в соответствии с оптимальным обходом городов
    for i in range(x.size):
        new_x[i] = x[int(optimal_path[i])]
        new_y[i] = y[int(optimal_path[i])]
    
    # Возвращаем обновленную последовательность координат
    return new_x, new_y

# Возвращает случайный набор точек на плоскости
def createDots(num_samples, box):
    x = np.random.random(num_samples)
    y = np.random.random(num_samples)
    x = x * (box[1] - box[0]) + box[0]
    y = y * (box[1] - box[0]) + box[0]
    return x, y

# Возвращает набор точек на окружности радиуса 'r' с центром в точке 'beg'
def createCircle(num_samples, r, beg):
    dot = np.zeros((num_samples, 2))
    alpha = 2 * 3.1415 / num_samples
    for i in range(num_samples):
        dot[i][0] = r * np.cos(i * alpha) + beg[0]
        dot[i][1] = r * np.sin(i * alpha) + beg[1]
    np.random.shuffle(dot)
    return dot[:, 0], dot[:, 1]

# Возвращает длину пути для цепочки городов
def length(x, y):
    cost = 0
    for i in range(x.size - 1):
        cost += np.sqrt((x[i] - x[i + 1]) ** 2 + (y[i] - y[i + 1]) ** 2)
    cost += np.sqrt((x[0] - x[x.size - 1]) ** 2 + (y[0] - y[y.size - 1]) ** 2)
    return cost

# Инициализация генератора случайных чисел 
np.random.seed(1)

# Генерируем случайный набор точек на плоскости
x, y = createDots(100, [-20, 20])
# x, y = createCircle(50, 20, [0, 0])
```

```python
# Выводим расположение городов на экран
fig, ax = plt.subplots()
fig.set_figwidth(6)
fig.set_figheight(6)
ax.set_title('Cities locations')
ax.set_xlim(-25, 25)
ax.set_ylim(-25, 25)
ax.scatter(x, y, s=5, color='black', label='Cities')
ax.grid(linestyle = '--')
ax.legend()
plt.show()
```

```python
# Выводим расположение городов на экран
fig, ax = plt.subplots()
fig.set_figwidth(6)
fig.set_figheight(6)
ax.set_title('Initial random path')
ax.set_xlim(-25, 25)
ax.set_ylim(-25, 25)
ax.scatter(x, y, s=5, color='black')
ax.plot(x, y, color='yellow', linestyle='--', label='Initial')
ax.grid(linestyle = '--')
ax.legend()
plt.show()

# Выводим длину маршрута
print('Total length: ', length(x, y))
```

```python
# Формируем оптимальный маршрут (муравьиная колония)
ant_x, ant_y = antColony(x, y, 500, 1, 7, 0.2)

# Выводим оптимальный путь
fig, ax = plt.subplots()
fig.set_figwidth(6)
fig.set_figheight(6)
ax.set_title('Optimal path (ant algorithm)')
ax.set_xlim(-25, 25)
ax.set_ylim(-25, 25)
ax.scatter(x, y, s=5, color='black')
ax.plot(ant_x, ant_y, color='yellow', linestyle='--', label='Optimal')
ax.grid(linestyle = '--')
ax.legend()
plt.show()

# Выводим длину маршрута
print('Total length: ', length(ant_x, ant_y))
```

```python
# Формируем оптимальный маршрут (метод имитации отжига)
sam_x, sam_y = simulatedAnnealing(x, y, 150, 1, 0.99)

# Выводим оптимальный путь
fig, ax = plt.subplots()
fig.set_figwidth(6)
fig.set_figheight(6)
ax.set_title('Optimal path (annealing simulation)')
ax.set_xlim(-25, 25)
ax.set_ylim(-25, 25)
ax.scatter(x, y, s=5, color='black')
ax.plot(sam_x, sam_y, color='yellow', linestyle='--', label='Optimal')
ax.grid(linestyle = '--')
ax.legend()
plt.show()

# Выводим длину маршрута
print('Total length: ', length(sam_x, sam_y))
```
