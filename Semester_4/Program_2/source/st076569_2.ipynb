```python
#################################################################################
# Дискретная классификация. Выбрать датасет, подходящий для классификации.      #
# Разделить датасет на тестовую/обучающую выборки. Построить дискретный         #
# классификатор на основе обучающей выборки, и проверить его качество на        #
# тестовой выборке (описать какими критериями пользовались).                    #
#################################################################################

#Импортируем нужные библиотеки:
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data
from pandas import DataFrame
%pylab inline

# Загружаем набор данных Ирисы
iris = datasets.load_iris()
iris_frame = DataFrame(iris.data)

# Делаем имена колонок такие же, как имена переменных
iris_frame.columns = iris.feature_names

# Добавляем столбец с целевой переменной
iris_frame['target'] = iris.target

# Для наглядности добавляем столбец с сортами
iris_frame['name'] = iris_frame.target.apply(lambda x : iris.target_names[x])

# Создаем обучающую и проверочную выборки
train_data, test_data, train_labels, test_labels = train_test_split(iris_frame[['sepal length (cm)',
                                                                                'sepal width (cm)',
                                                                                'petal length (cm)',
                                                                                'petal width (cm)']], 
                                                                    iris_frame['target'], 
                                                                    test_size = 0.5, 
                                                                    random_state = 0)

# Строим дерево классификации
model = DecisionTreeClassifier(criterion='gini', max_depth=8, random_state=0)
model.fit(train_data, train_labels)

# Проводим тестирование на тестовой выборке
model_predictions = model.predict(test_data)

# Выводим инфомацию о точности классификатора
print('\n', metrics.classification_report(test_labels, model_predictions, target_names=iris.target_names))

# Визуализируем дерево
dot_data = export_graphviz(model,
                           filled=True, 
                           rounded=True,
                           class_names=iris.target_names,
                           feature_names=iris.feature_names,
                           out_file=None) 

# Сохраняем рисунок дерева
graph = graph_from_dot_data(dot_data)
graph.write_png('tree.png')
```
